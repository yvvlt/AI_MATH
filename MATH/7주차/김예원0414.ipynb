```python
import numpy as np

A = np.array([[1, 3],
              [3, 1]])

eigenvalues, eigenvectors = np.linalg.eig(A)

print("고유값:", eigenvalues)
print("고유벡터:\n", eigenvectors)
```

```python
from scipy.optimize import minimize

# 함수 정의
def L(params):
    w, b = params
    return 0.5 * ((1 - w - b)**2 + b**2)

# 초기값
initial = [-1, 1]

# 최적화 수행
result = minimize(L, initial)

print("최솟값을 갖는 (w, b):", result.x)
```
